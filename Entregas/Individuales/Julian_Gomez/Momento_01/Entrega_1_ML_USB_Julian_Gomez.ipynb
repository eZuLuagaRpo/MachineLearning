### **Ejercicio 1 (C = 0.7)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/1Z3yNgkQVFJyaCRyTzkjFhIB_3kxqs61q/view?usp=drive_link). \\
**NOTA:** En la salida de la celda puede encontrar el tamaño de la imagen final.

```
import numpy as np
import matplotlib.pyplot as plt

# Tamaño de la imagen
h, w = 512, 512

# Crear gradiente horizontal de 0 a 200 (gris)
gradient = np.linspace(0, 200, w)

# Repetir el gradiente en todas las filas
img1 = np.tile(gradient, (h, 1))

# Convertir a uint8
img1 = img1.astype(np.uint8)

# Mostrar
plt.imshow(img1, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

# Guardar
plt.imsave("imagen_gradiente.png", img1, cmap="gray")

#Se genera con: I(x,y)=x, y se normaliza al rango 0–255.
```

### **Ejercicio 2 (C = 1.0)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/13z0tvGYD-37_PuHOqoRWHGGmGq7xxvU7/view?usp=drive_link).

```
import numpy as np
import matplotlib.pyplot as plt

# Tamaño
h, w = 512, 512

# Crear índices de la matriz
y, x = np.indices((h, w))

# Generar patrón diagonal
pattern = (x - y) % 40   # controla el ancho del patrón

# Crear franjas
img2 = np.where(pattern < 20, 255, 0)

img2 = img2.astype(np.uint8)

# Mostrar
plt.imshow(img2, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()

# Guardar
plt.imsave("imagen_diagonal.png", img2, cmap="gray")

#Se usa: I(x,y)=(x+y) mod p donde p controla el espaciado de las líneas.
```

### **Ejercicio 3 (C = 1.6)**

Genere la siguiente animación [resultado](https://drive.google.com/file/d/1jSNhV2CYdm6-E5wuo7o5wE-b2J-_mEil/view?usp=drive_link).

```
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread, imwrite

lena = imread("/content/lena.png")
tiger = imread("/content/bengal_tiger.jpg")

h,w,c = lena.shape

resultado = np.zeros_like(lena)

bloque = 1

for i in range(h):
    for j in range(w):

        bi = (i // bloque) % 2
        bj = (j // bloque) % 2

        if (bi + bj) % 2 == 0:
            resultado[i,j] = lena[i,j]
        else:
            resultado[i,j] = tiger[i,j]

plt.imshow(resultado)
plt.axis("off")

imwrite("resultado_combinado.jpg", resultado)
```

### **Ejercicio 4 (C = 1.0)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/1UX2Y8Q2xWy5ej0hl_r1B-fKAQ8SX5DI1/view?usp=drive_link).

```
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread, imwrite

# cargar imagen
img = imread("/content/lena.png")

h, w, c = img.shape

resultado = np.zeros_like(img)

for i in range(h):
    for j in range(w):

        if i < h//2:
            resultado[i,j] = img[i,j]
        else:
            resultado[i,j] = img[h-i-1,j]

plt.imshow(resultado)
plt.axis("off")

imwrite("resultado_espejo.jpg", resultado)
```

### **Ejercicio 5 (C = 0.7)**

Genere la siguiente imagen [resultado](https://drive.google.com/file/d/10jxoa_muSn6bxsDBy7Vw88wz1oirMiRH/view?usp=drive_link)

```
import numpy as np
import matplotlib.pyplot as plt
from imageio import imread, imwrite

img = imread("/content/lena.png")

h,w,c = img.shape

resultado = np.zeros_like(img)

bloque = 60

for i in range(h):
    for j in range(w):

        bi = (i // bloque) % 3
        bj = (j // bloque) % 3

        if (bi - bj) % 3 == 0:
            resultado[i,j] = [img[i,j,0],0,0]   # rojo
        elif (bi - bj) % 3 == 1:
            resultado[i,j] = [0,img[i,j,1],0]   # verde
        else:
            resultado[i,j] = [0,0,img[i,j,2]]   # azul

plt.imshow(resultado)
plt.axis("off")

imwrite("resultado_mosaico.jpg", resultado)
```
