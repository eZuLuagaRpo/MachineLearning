## Objetivo
Comprender cómo funciona Gradient Descent para minimizar una función de costo y entrenar un modelo simple de regresión lineal.


## Bloque 1. Importar librerías

En este bloque cargamos librerías para cálculo numérico y visualización.


```python
# Importamos NumPy para operaciones numéricas.
import numpy as np

# Importamos Matplotlib para gráficas.
import matplotlib.pyplot as plt
```

### Interpretación del bloque 1

Este bloque deja listo el entorno para trabajar con optimización y visualización de funciones de costo.


## Bloque 2. Crear datos sintéticos

Vamos a crear un conjunto de datos simple con una relación aproximadamente lineal.


```python
# Definimos semilla para reproducibilidad.
np.random.seed(42)

# Generamos valores de entrada.
X = 2 * np.random.rand(100, 1)

# Generamos valores de salida con ruido.
y = 4 + 3 * X + np.random.randn(100, 1)

# Mostramos las primeras observaciones.
print(X[:5])
print(y[:5])

# Graficamos los datos.
plt.scatter(X, y)
plt.title("Datos sintéticos")
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

### Interpretación del bloque 2

Los datos muestran una tendencia lineal aproximada.

**Qué significa:**
- Existe una relación entre X e y.
- Gradient Descent intentará encontrar la mejor línea que minimice el error.


## Bloque 3. Inicializar parámetros del modelo

Aquí definimos pesos iniciales y parámetros de entrenamiento.


```python
# Número de observaciones.
m = len(X)

# Inicializamos theta0 y theta1 aleatoriamente.
theta0 = np.random.randn()
theta1 = np.random.randn()

# Definimos learning rate.
learning_rate = 0.1

# Número de iteraciones.
iterations = 100

# Lista para guardar costo.
cost_history = []
```

### Interpretación del bloque 3

Aquí el modelo todavía no sabe nada.

**Qué significa:**
- Los parámetros iniciales son aleatorios.
- Gradient Descent irá ajustándolos poco a poco para reducir el error.


## Bloque 4. Aplicar Gradient Descent

En este bloque entrenamos el modelo iterativamente.


```python
# Recorremos el número de iteraciones.
for i in range(iterations):

    # Calculamos predicciones actuales.
    y_pred = theta0 + theta1 * X

    # Calculamos error.
    error = y_pred - y

    # Calculamos gradientes.
    d_theta0 = (2/m) * np.sum(error)
    d_theta1 = (2/m) * np.sum(error * X)

    # Actualizamos parámetros.
    theta0 = theta0 - learning_rate * d_theta0
    theta1 = theta1 - learning_rate * d_theta1

    # Calculamos costo MSE.
    cost = (1/m) * np.sum(error**2)

    # Guardamos costo.
    cost_history.append(cost)

# Mostramos parámetros finales.
print("theta0:", theta0)
print("theta1:", theta1)
```

### Interpretación del bloque 4

Gradient Descent ajustó progresivamente los parámetros para reducir el error.

**Qué ocurre realmente:**
- En cada iteración el algoritmo calcula en qué dirección debe moverse.
- Los gradientes indican cómo cambiar los parámetros para disminuir el costo.
- El learning rate controla el tamaño del paso.


## Bloque 5. Visualizar convergencia del costo

En este bloque veremos cómo disminuye el error a medida que avanzan las iteraciones.


```python
# Graficamos la evolución del costo.
plt.plot(range(iterations), cost_history)

# Agregamos título y etiquetas.
plt.title("Convergencia del costo")
plt.xlabel("Iteraciones")
plt.ylabel("Costo")

# Mostramos gráfica.
plt.show()
```

### Interpretación del bloque 5

La gráfica muestra cómo el costo disminuye con las iteraciones.

**Qué significa:**
- Si el costo baja, el modelo está aprendiendo.
- Si el costo diverge, el learning rate probablemente es demasiado grande.


## Bloque 6. Visualizar la recta final

Aquí mostramos la línea aprendida por el modelo.


```python
# Calculamos predicciones finales.
y_final = theta0 + theta1 * X

# Graficamos datos originales.
plt.scatter(X, y)

# Graficamos la recta aprendida.
plt.plot(X, y_final, color="red")

# Agregamos título.
plt.title("Regresión lineal con Gradient Descent")

# Mostramos gráfica.
plt.show()
```

### Interpretación del bloque 6

La línea roja representa el modelo aprendido.

**Idea clave:**
Gradient Descent encontró parámetros que aproximan correctamente la relación entre los datos.
