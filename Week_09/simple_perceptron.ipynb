## Objetivo
Comprender el funcionamiento básico del perceptrón como uno de los modelos fundamentales de redes neuronales.


## Bloque 1. Importar librerías

En este bloque cargamos librerías para cálculo y visualización.


```python
# Importamos NumPy para cálculos numéricos.
import numpy as np

# Importamos Matplotlib para gráficas.
import matplotlib.pyplot as plt

# Importamos Perceptron desde scikit-learn.
from sklearn.linear_model import Perceptron
```

### Interpretación del bloque 1

Este bloque deja listo el entorno para trabajar con un clasificador lineal inspirado en neuronas artificiales.


## Bloque 2. Crear dataset binario

En este bloque vamos a generar datos simples de clasificación binaria.


```python
# Definimos manualmente puntos de entrada.
X = np.array([
    [1, 1],
    [2, 1],
    [2, 2],
    [3, 2],
    [6, 5],
    [7, 7],
    [8, 6],
    [9, 8]
])

# Definimos clases binarias.
y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

# Graficamos los datos.
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm", edgecolors="k")

# Agregamos título.
plt.title("Dataset binario")

# Mostramos gráfica.
plt.show()
```

### Interpretación del bloque 2

La gráfica muestra dos clases relativamente separables.

**Qué significa:**
- El perceptrón intentará encontrar una frontera lineal para separar ambas clases.


## Bloque 3. Entrenar el perceptrón

En este bloque vamos a entrenar el modelo.


```python
# Creamos el modelo perceptrón.
model = Perceptron(
    max_iter=1000,      # Máximo número de iteraciones
    eta0=0.1,           # Learning rate
    random_state=42     # Reproducibilidad
)

# Entrenamos el modelo.
model.fit(X, y)

# Mostramos pesos aprendidos.
print("Pesos:", model.coef_)

# Mostramos bias.
print("Bias:", model.intercept_)
```

### Interpretación del bloque 3

El perceptrón ajustó pesos para construir una frontera lineal.

**Qué significa:**
- Los pesos indican la importancia de cada variable.
- El bias desplaza la frontera.
- El modelo aprende corrigiendo errores de clasificación.


## Bloque 4. Hacer predicciones

En este bloque probamos el modelo con nuevos puntos.


```python
# Creamos nuevos puntos.
new_points = np.array([
    [2, 2],
    [8, 7],
    [4, 3]
])

# Predecimos clases.
predictions = model.predict(new_points)

# Mostramos resultados.
for point, pred in zip(new_points, predictions):
    print(f"Punto {point} → Clase {pred}")
```

### Interpretación del bloque 4

Aquí el perceptrón toma decisiones sobre datos nuevos.

**Qué significa:**
- El modelo ya construyó una frontera de separación.
- Ahora puede clasificar observaciones no vistas previamente.


## Bloque 5. Visualizar frontera de decisión

En este bloque dibujamos la frontera aprendida por el perceptrón.


```python
# Extraemos pesos.
w = model.coef_[0]

# Extraemos bias.
b = model.intercept_[0]

# Generamos valores para la línea.
x_vals = np.linspace(0, 10, 100)

# Calculamos y según la ecuación de la recta.
y_vals = -(w[0] * x_vals + b) / w[1]

# Graficamos datos.
plt.scatter(X[:,0], X[:,1], c=y, cmap="coolwarm", edgecolors="k")

# Graficamos frontera.
plt.plot(x_vals, y_vals, color="black")

# Agregamos título.
plt.title("Frontera de decisión del perceptrón")

# Mostramos gráfica.
plt.show()
```

### Interpretación del bloque 5

La línea negra representa la frontera de decisión.

**Qué significa:**
- Un lado de la línea pertenece a la clase 0.
- El otro lado pertenece a la clase 1.
- El perceptrón funciona bien cuando las clases son linealmente separables.

### Ideas clave

- Aprende ajustando pesos y bias.
- Construye fronteras lineales.
- Es la base conceptual de redes neuronales más complejas.
