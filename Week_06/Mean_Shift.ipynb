Este notebook sigue la misma estructura pedagógica:
1. Antes del bloque: qué se va a hacer.

3. Después del bloque: interpretación.

## Objetivo
Comprender Mean Shift y compararlo con K-Means en datos sintéticos, datos reales e imágenes.


## Bloque 1. Importar librerías

En este bloque vamos a cargar librerías para generar datos, entrenar Mean Shift y visualizar resultados.


```python
# Importamos NumPy.
import numpy as np

# Importamos Pandas.
import pandas as pd

# Importamos MeanShift.
from sklearn.cluster import MeanShift

# Importamos make_blobs.
from sklearn.datasets import make_blobs

# Importamos Matplotlib.
from matplotlib import pyplot as plt
```

### Interpretación del bloque 1

Mean Shift pertenece al aprendizaje no supervisado, pero a diferencia de K-Means, no exige definir `k` explícitamente.


## Bloque 2. Generar datos sintéticos y visualizarlos

En este bloque creamos un dataset artificial para observar cómo Mean Shift encuentra grupos por densidad.


```python
# Definimos número de muestras.
n_samples = 300

# Definimos semilla aleatoria.
random_state = 42

# Generamos datos sintéticos con cuatro grupos.
X, Y = make_blobs(
    n_samples=n_samples,
    centers=4,
    cluster_std=1.0,
    random_state=random_state
)

# Creamos una figura con dos subgráficas.
_, axes = plt.subplots(1, 2, figsize=(13, 5))

# Mostramos los datos sin etiquetas visibles.
axes[0].scatter(X[:, 0], X[:, 1], edgecolors="k")
axes[0].set_title("Datos sin etiquetas")

# Mostramos los datos con grupos reales.
axes[1].scatter(X[:, 0], X[:, 1], c=Y, cmap="viridis", edgecolors="k")
axes[1].set_title("Datos con grupos reales")

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 2

La primera gráfica representa lo que realmente ve el algoritmo: solo puntos. La segunda sirve como referencia para comparar si Mean Shift logra encontrar una estructura similar.


## Bloque 3. Entrenar Mean Shift en datos sintéticos

En este bloque entrenamos Mean Shift, obtenemos centros y etiquetas, y visualizamos el resultado.


```python
# Creamos el modelo Mean Shift.
ms = MeanShift()

# Entrenamos el modelo con los datos.
ms.fit(X)

# Guardamos los centros encontrados.
cluster_centers = ms.cluster_centers_

# Guardamos las etiquetas predichas.
labels = ms.labels_

# Mostramos los centros.
print(cluster_centers)

# Mostramos cuántos clusters encontró.
print("Número de clusters encontrados:", len(cluster_centers))

# Creamos la figura.
plt.figure(figsize=(7, 5))

# Dibujamos los puntos según cluster encontrado.
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap="viridis", edgecolors="k")

# Dibujamos los centros.
plt.scatter(cluster_centers[:, 0], cluster_centers[:, 1], c="red", s=250, marker="X", label="Centros")

# Agregamos título y leyenda.
plt.title("Mean Shift sobre datos sintéticos")
plt.legend()

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 3

Mean Shift encuentra grupos automáticamente buscando regiones densas.

**Diferencia clave con K-Means:**
- K-Means necesita que le digamos cuántos grupos buscar.
- Mean Shift descubre automáticamente cuántas concentraciones relevantes existen.


## Bloque 4. Aplicar Mean Shift sobre Iris

En este bloque usamos el dataset Iris para observar el comportamiento del algoritmo en un caso real.


```python
# Importamos Iris.
from sklearn.datasets import load_iris

# Cargamos el dataset.
data = load_iris()

# Guardamos las variables.
x = data.data

# Guardamos las clases reales como referencia.
y = data.target

# Creamos una gráfica de referencia.
plt.figure(figsize=(7, 5))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap=plt.cm.Paired)
plt.title("Iris - referencia visual")
print(x.shape, y.shape)
plt.show()
```

### Interpretación del bloque 4

Este bloque nos permite comparar clustering sobre un dataset clásico, no solo sobre ejemplos sintéticos controlados.


## Bloque 5. Entrenar Mean Shift en Iris y visualizar

En este bloque entrenamos el modelo sobre Iris y observamos cuántos grupos encuentra.


```python
# Creamos el modelo Mean Shift.
ms = MeanShift()

# Entrenamos el modelo.
ms.fit(x)

# Guardamos centroides.
centroids = ms.cluster_centers_

# Guardamos etiquetas predichas.
predictions = ms.labels_

# Mostramos los centroides.
print(centroids)

# Mostramos número de clusters.
print("Número de clusters encontrados:", len(centroids))

# Creamos una figura.
plt.figure(figsize=(7, 5))

# Dibujamos los puntos según cluster encontrado.
plt.scatter(x[:, 0], x[:, 1], c=predictions, cmap="viridis", edgecolors="k")

# Dibujamos los centroides.
plt.scatter(centroids[:, 0], centroids[:, 1], c="red", s=250, marker="X", label="Centroides")

# Agregamos título y leyenda.
plt.title("Mean Shift en Iris")
plt.legend()

# Mostramos la figura.
plt.show()
```

### Interpretación del bloque 5

El número de clusters encontrados puede no coincidir exactamente con las clases reales.

**Qué enseña esto:**
Clustering descubre estructura estadística, no necesariamente categorías humanas conocidas.


## Bloque 6. Aplicar Mean Shift a una imagen

En este bloque tratamos una imagen como un conjunto de píxeles para segmentarla con Mean Shift.


```python
# Importamos una imagen de ejemplo.
from sklearn.datasets import load_sample_image

# Importamos conversión RGB a HSV.
from matplotlib.colors import rgb_to_hsv

# Cargamos la imagen flower.jpg.
original_image = load_sample_image("flower.jpg")

# Reducimos el tamaño de la imagen.
original_image = original_image[::4, ::4]

# Normalizamos a valores entre 0 y 1.
img_rgb = original_image / 255.0

# Convertimos a HSV.
img_hsv = rgb_to_hsv(img_rgb)

# Mostramos RGB y HSV.
_, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_rgb)
axes[0].set_axis_off()
axes[0].set_title("Imagen RGB")
axes[1].imshow(img_hsv)
axes[1].set_axis_off()
axes[1].set_title("Imagen HSV")
plt.show()

# Vectorizamos RGB y HSV.
vectorized_rgb = np.float32(img_rgb.reshape((-1, 3)))
vectorized_hsv = np.float32(img_hsv.reshape((-1, 3)))

# Mostramos dimensiones.
print(vectorized_rgb.shape, vectorized_hsv.shape)
```

### Interpretación del bloque 6

Una imagen puede verse como una tabla de datos donde cada fila es un píxel y cada columna un canal de color. Eso permite aplicar clustering igual que en un dataset tabular.


## Bloque 7. Segmentar en RGB con Mean Shift

En este bloque segmentamos la imagen en el espacio RGB.


```python
# Entrenamos Mean Shift con los píxeles RGB.
ms_rgb = MeanShift(max_iter=5).fit(vectorized_rgb)

# Recuperamos etiquetas.
clustered_rgb = ms_rgb.labels_

# Reorganizamos con la forma de la imagen.
clustered_rgb = clustered_rgb.reshape(img_rgb.shape[:2])

# Mostramos dimensiones.
print(vectorized_rgb.shape, img_rgb.shape, clustered_rgb.shape)

# Creamos figura comparativa.
_, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_rgb)
axes[0].set_axis_off()
axes[0].set_title("Original")
axes[1].imshow(clustered_rgb, cmap="Paired")
axes[1].set_axis_off()
axes[1].set_title("Segmentación RGB")
plt.show()

# Elegimos un cluster para aislar.
cluster = 0

# Creamos una máscara binaria.
cluster_mask = np.array(clustered_rgb == cluster, dtype=np.uint8)

# Copiamos la imagen.
masked_image = np.copy(img_rgb)

# Ponemos en negro lo que no pertenece al cluster.
masked_image[cluster_mask == 0] = 0

# Mostramos la máscara.
plt.figure(figsize=(6, 5))
plt.imshow(masked_image)
plt.axis("off")
plt.title("Cluster aislado en RGB")
plt.show()
```

### Interpretación del bloque 7

Mean Shift segmenta automáticamente regiones densas de color. Esto lo vuelve interesante cuando no se quiere fijar previamente el número de grupos.


## Bloque 8. Segmentar en HSV y comparar

En este bloque repetimos la segmentación en HSV para analizar el efecto del espacio de color.


```python
# Entrenamos Mean Shift con los píxeles HSV.
ms_hsv = MeanShift(max_iter=5).fit(vectorized_hsv)

# Recuperamos etiquetas.
clustered_hsv = ms_hsv.labels_

# Reorganizamos con la forma de la imagen.
clustered_hsv = clustered_hsv.reshape(img_rgb.shape[:2])

# Mostramos dimensiones.
print(vectorized_hsv.shape, img_rgb.shape, clustered_hsv.shape)

# Creamos figura comparativa.
_, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].imshow(img_rgb)
axes[0].set_axis_off()
axes[0].set_title("Original")
axes[1].imshow(clustered_hsv, cmap="Paired")
axes[1].set_axis_off()
axes[1].set_title("Segmentación HSV")
plt.show()

# Elegimos un cluster.
cluster = 1

# Creamos máscara.
cluster_mask = np.array(clustered_hsv == cluster, dtype=np.uint8)

# Copiamos la imagen original.
masked_image = np.copy(img_rgb)

# Ponemos en negro lo que no pertenece al cluster.
masked_image[cluster_mask == 0] = 0

# Mostramos máscara.
plt.figure(figsize=(6, 5))
plt.imshow(masked_image)
plt.axis("off")
plt.title("Cluster aislado en HSV")
plt.show()
```

### Interpretación del bloque 8

La comparación entre RGB y HSV muestra que la representación de los datos influye directamente en el resultado del clustering.

**Mensaje docente:**
La calidad de la segmentación depende tanto del algoritmo como del espacio donde representamos la información.

- No requiere definir `k` de forma explícita.
