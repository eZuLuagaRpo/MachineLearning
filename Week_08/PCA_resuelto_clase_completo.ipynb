# Taller resuelto: PCA - Análisis de Componentes Principales

## Objetivo
Aplicar PCA para reducir dimensionalidad, visualizar datos y entender varianza explicada.

## Bloque 1. Importar librerías

Cargaremos herramientas para datos, escalado, PCA y visualización.

```python
# Importamos NumPy.
import numpy as np

# Importamos Pandas.
import pandas as pd

# Importamos Matplotlib.
import matplotlib.pyplot as plt

# Importamos dataset Iris.
from sklearn.datasets import load_iris

# Importamos escalador.
from sklearn.preprocessing import StandardScaler

# Importamos PCA.
from sklearn.decomposition import PCA
```

### Interpretación

PCA trabaja con varianza; por eso normalmente requiere escalar antes de transformar los datos.

## Bloque 2. Cargar datos

Usaremos Iris para reducir de 4 variables a 2 componentes principales.

```python
# Cargamos Iris.
iris = load_iris()

# Guardamos variables.
X = iris.data

# Guardamos etiquetas para visualización.
y = iris.target

# Creamos DataFrame.
df = pd.DataFrame(X, columns=iris.feature_names)

# Mostramos primeras filas.
display(df.head())

# Mostramos dimensiones.
print("Shape de X:", X.shape)
```

### Interpretación

El dataset tiene 4 variables numéricas. PCA permitirá resumirlas en menos dimensiones.

## Bloque 3. Escalar datos

Estandarizaremos las variables para que tengan media 0 y desviación estándar 1.

```python
# Creamos escalador.
scaler = StandardScaler()

# Ajustamos y transformamos datos.
X_scaled = scaler.fit_transform(X)

# Mostramos primeras filas escaladas.
print(X_scaled[:5])
```

### Interpretación

Escalar evita que una variable domine PCA solo por tener una escala numérica mayor.

## Bloque 4. Aplicar PCA con dos componentes

Reduciremos las 4 variables originales a 2 componentes para poder visualizar.

```python
# Creamos PCA con dos componentes.
pca = PCA(n_components=2)

# Ajustamos y transformamos.
X_pca = pca.fit_transform(X_scaled)

# Mostramos dimensiones originales.
print("Dimensión original:", X_scaled.shape)

# Mostramos dimensiones reducidas.
print("Dimensión reducida:", X_pca.shape)

# Mostramos primeras filas transformadas.
print(X_pca[:5])
```

### Interpretación

PCA crea nuevas variables llamadas componentes principales. No son columnas originales, sino combinaciones lineales que concentran varianza.

## Bloque 5. Visualizar en espacio PCA

Graficaremos los dos primeros componentes principales.

```python
# Creamos figura.
plt.figure(figsize=(7,5))

# Dibujamos los datos en espacio PCA.
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap="viridis", edgecolors="k")

# Agregamos etiquetas.
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")

# Agregamos título.
plt.title("Iris proyectado en 2 componentes PCA")

# Mostramos gráfica.
plt.show()
```

### Interpretación

La gráfica permite ver si los grupos mantienen separación en menos dimensiones. PCA no usa etiquetas, solo estructura de varianza.

## Bloque 6. Varianza explicada

Analizaremos cuánta información conserva cada componente.

```python
# Obtenemos proporción de varianza explicada.
explained = pca.explained_variance_ratio_

# Mostramos varianza por componente.
print("Varianza explicada:", explained)

# Mostramos varianza acumulada.
print("Varianza acumulada:", np.cumsum(explained))

# Creamos gráfica.
plt.figure(figsize=(7,5))

# Dibujamos barras.
plt.bar(["PC1", "PC2"], explained)

# Agregamos título.
plt.title("Varianza explicada por componente")
plt.ylabel("Proporción de varianza")

# Mostramos gráfica.
plt.show()
```

### Interpretación

Si dos componentes explican gran parte de la varianza, la reducción a 2D conserva buena parte de la información original.

## Bloque 7. PCA con todos los componentes

Evaluaremos cuántos componentes se necesitan para conservar suficiente varianza.

```python
# Creamos PCA completo.
pca_full = PCA()

# Ajustamos PCA completo.
pca_full.fit(X_scaled)

# Calculamos varianza acumulada.
cum_var = np.cumsum(pca_full.explained_variance_ratio_)

# Creamos figura.
plt.figure(figsize=(7,5))

# Dibujamos varianza acumulada.
plt.plot(range(1, len(cum_var)+1), cum_var, marker="o")

# Agregamos línea de referencia al 95%.
plt.axhline(0.95, linestyle="--", color="red", label="95%")

# Agregamos título y etiquetas.
plt.title("Varianza explicada acumulada")
plt.xlabel("Número de componentes")
plt.ylabel("Varianza acumulada")
plt.legend()

# Mostramos gráfica.
plt.show()
```

### Interpretación

Esta curva ayuda a decidir cuántos componentes usar. Se busca conservar alta varianza usando la menor cantidad posible de componentes.

## Bloque 8. Reconstrucción conceptual

Reconstruiremos una aproximación de los datos escalados desde los dos componentes.

```python
# Reconstruimos datos desde los dos componentes.
X_reconstructed = pca.inverse_transform(X_pca)

# Mostramos dimensión reconstruida.
print("Dimensión reconstruida:", X_reconstructed.shape)

# Mostramos primeras filas.
print(X_reconstructed[:5])
```

### Interpretación

La reconstrucción muestra que PCA comprime y aproxima. Si la varianza explicada es alta, la pérdida de información será menor.

## Cierre

PCA es útil para visualización, reducción de dimensionalidad, compresión y preprocesamiento antes de otros modelos.
