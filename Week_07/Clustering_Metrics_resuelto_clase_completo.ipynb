## Objetivo
Evaluar modelos de clustering usando métricas internas y externas: Silhouette, Calinski-Harabasz, Davies-Bouldin, ARI y NMI.

## Bloque 1. Importar librerías

En este bloque importaremos herramientas para generar datos, entrenar K-Means, graficar y calcular métricas de clustering.

```python
# Importamos NumPy para manejar arreglos numéricos.
import numpy as np

# Importamos Pandas para organizar resultados en tablas.
import pandas as pd

# Importamos Matplotlib para crear visualizaciones.
import matplotlib.pyplot as plt

# Importamos make_blobs para crear datos sintéticos con grupos conocidos.
from sklearn.datasets import make_blobs

# Importamos KMeans como algoritmo de clustering base.
from sklearn.cluster import KMeans

# Importamos métricas internas y externas de clustering.
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score, adjusted_rand_score, normalized_mutual_info_score
```

### Interpretación del bloque 1

Este bloque prepara el entorno. Las métricas internas no necesitan etiquetas reales; las externas sí comparan contra una referencia conocida.

## Bloque 2. Generar datos sintéticos

Crearemos un dataset con 4 grupos reales para poder evaluar si el clustering recupera la estructura esperada.

```python
# Generamos datos sintéticos con 4 centros.
X, y_true = make_blobs(n_samples=600, centers=4, cluster_std=1.1, random_state=42)

# Mostramos la forma de la matriz de características.
print("Shape de X:", X.shape)

# Mostramos la forma de las etiquetas reales.
print("Shape de y_true:", y_true.shape)

# Creamos una figura.
plt.figure(figsize=(7,5))

# Graficamos los datos con el color de la clase real.
plt.scatter(X[:,0], X[:,1], c=y_true, cmap="viridis", edgecolors="k")

# Agregamos título y etiquetas.
plt.title("Datos sintéticos con grupos reales")
plt.xlabel("Variable 1")
plt.ylabel("Variable 2")

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 2

La gráfica permite ver cuatro grupos relativamente compactos. Este escenario es adecuado para practicar evaluación de clustering porque conocemos la estructura real.

## Bloque 3. Entrenar K-Means para varios valores de K

Entrenaremos K-Means con diferentes valores de `k` y calcularemos métricas internas y externas para cada caso.

```python
# Definimos los valores de k que vamos a evaluar.
k_values = range(2, 9)

# Creamos listas para almacenar las métricas.
silhouette_scores = []
calinski_scores = []
davies_scores = []
ari_scores = []
nmi_scores = []

# Recorremos cada valor de k.
for k in k_values:
    # Creamos el modelo K-Means.
    model = KMeans(n_clusters=k, n_init=10, random_state=42)
    
    # Entrenamos el modelo y obtenemos las etiquetas predichas.
    y_pred = model.fit_predict(X)
    
    # Calculamos Silhouette Score.
    silhouette_scores.append(silhouette_score(X, y_pred))
    
    # Calculamos Calinski-Harabasz Index.
    calinski_scores.append(calinski_harabasz_score(X, y_pred))
    
    # Calculamos Davies-Bouldin Index.
    davies_scores.append(davies_bouldin_score(X, y_pred))
    
    # Calculamos Adjusted Rand Index.
    ari_scores.append(adjusted_rand_score(y_true, y_pred))
    
    # Calculamos Normalized Mutual Information.
    nmi_scores.append(normalized_mutual_info_score(y_true, y_pred))

# Confirmamos que el proceso terminó.
print("Modelos evaluados correctamente")
```

### Interpretación del bloque 3

Cada valor de `k` produce una partición diferente. Las métricas permiten comparar esas particiones y justificar una elección.

## Bloque 4. Visualizar métricas internas

Graficaremos Silhouette, Calinski-Harabasz y Davies-Bouldin para comparar la calidad interna de los clusters.

```python
# Creamos una figura para Silhouette Score.
plt.figure(figsize=(8,5))
plt.plot(list(k_values), silhouette_scores, marker="o")
plt.title("Silhouette Score por K")
plt.xlabel("Número de clusters")
plt.ylabel("Silhouette")
plt.show()

# Creamos una figura para Calinski-Harabasz.
plt.figure(figsize=(8,5))
plt.plot(list(k_values), calinski_scores, marker="o")
plt.title("Calinski-Harabasz por K")
plt.xlabel("Número de clusters")
plt.ylabel("Calinski-Harabasz")
plt.show()

# Creamos una figura para Davies-Bouldin.
plt.figure(figsize=(8,5))
plt.plot(list(k_values), davies_scores, marker="o")
plt.title("Davies-Bouldin por K")
plt.xlabel("Número de clusters")
plt.ylabel("Davies-Bouldin")
plt.show()
```

### Interpretación del bloque 4

Silhouette y Calinski-Harabasz se prefieren altos. Davies-Bouldin se prefiere bajo. Si varias métricas favorecen el mismo K, la decisión es más sólida.

## Bloque 5. Métricas externas y tabla resumen

Organizaremos todas las métricas en una tabla para comparar de manera integrada.

```python
# Construimos una tabla con todos los resultados.
results = pd.DataFrame({
    "k": list(k_values),
    "Silhouette": silhouette_scores,
    "Calinski_Harabasz": calinski_scores,
    "Davies_Bouldin": davies_scores,
    "ARI": ari_scores,
    "NMI": nmi_scores
})

# Mostramos la tabla.
results
```

### Interpretación del bloque 5

ARI y NMI comparan contra etiquetas reales; por eso son útiles en ejercicios controlados. En casos reales, normalmente se decide con métricas internas, visualización y criterio de negocio.

## Bloque 6. Entrenar el modelo final

Usaremos `k=4` como modelo final y visualizaremos los clusters y centroides.

```python
# Definimos el número de clusters final.
best_k = 4

# Creamos el modelo final.
best_model = KMeans(n_clusters=best_k, n_init=10, random_state=42)

# Entrenamos el modelo y obtenemos etiquetas.
best_labels = best_model.fit_predict(X)

# Guardamos los centroides.
centroids = best_model.cluster_centers_

# Creamos una figura.
plt.figure(figsize=(7,5))

# Dibujamos los puntos coloreados por cluster.
plt.scatter(X[:,0], X[:,1], c=best_labels, cmap="viridis", edgecolors="k")

# Dibujamos los centroides.
plt.scatter(centroids[:,0], centroids[:,1], c="red", s=250, marker="X", label="Centroides")

# Agregamos título y leyenda.
plt.title("Modelo final K-Means con k=4")
plt.legend()

# Mostramos la gráfica.
plt.show()
```

### Interpretación del bloque 6

El modelo final recupera cuatro regiones principales. La elección de K queda respaldada por visualización y métricas.
